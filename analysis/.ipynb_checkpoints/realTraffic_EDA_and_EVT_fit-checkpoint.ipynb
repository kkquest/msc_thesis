```python
# Define the post-gap start time
post_gap_start = pd.Timestamp("2015-09-09 00:00:00")

# Filter
df_post = df[df["timestamp"] >= post_gap_start].reset_index(drop=True)
clean_data = df_post[["timestamp", "value"]].copy()
clean_data.to_csv("../data/clean_data/post_gap_occupancy.csv", index=False)
```

#### Now let's explore the type of drift in the dataset.

In order to detect the presence of the drift, we can use rolling statistics.

Rolling statistics let us see how the baseline (mean) and volatility (std) change over time, which is exactly what “concept drift” means.

#### What rolling window to choose?

The dataset is recorded every 5 minutes. One day has 24h × 60min / 5min = 288 samples. So let's choose 144 (half of 288) as a compromise window to show both intra-day trends (≈12 hours) and to get a smoother estimate without waiting a full day.

```python
# Rolling statistics to visualise drift
window = 144
clean_data["roll_mean"] = clean_data["value"].rolling(window).mean()
clean_data["roll_std"]  = clean_data["value"].rolling(window).std()

# Plot the results
plt.figure(figsize=(10,3))
plt.plot(clean_data["timestamp"], clean_data["roll_mean"], label="Rolling Mean")
plt.plot(clean_data["timestamp"], clean_data["roll_std"],  label="Rolling Std")
plt.legend()
plt.title(f"Rolling Window={window} Stats")
plt.tight_layout()
plt.show()
```

```python
# Check the value distribution
plt.figure(figsize=(6,2.5))
plt.hist(clean_data["value"], bins=30, edgecolor="k")
plt.title("Value Distribution")
plt.tight_layout()
plt.show()
```

#### Main observations:
1. Rolling mean swings between 3 and 6 over the series.
2. Rolling STD hovers 2 and 5, with peaks corresponding to high-variance periods.
3. The Value Distribution shows quite heavy right tail
   
#### Implication:
both mean and variance drift over time (not purely stationary).

Now, let'c check the ACF 

```python
value = clean_data["value"].values

plt.figure(figsize=(8, 3))
plot_acf(value, lags=600)
plt.title("Autocorrelation Function (ACF)")
plt.tight_layout()
plt.show()
```

????? Peaks and troughs recur approximately every ~144 lags, which corresponds to one day of data if you're using 5-minute intervals (144 × 5min = 720min = 12 hours, so cycle = half-day or day)

### 2. EVT Suitabilty Check

```python
# Plot Empirical CDF 
values = clean_data['value'].values
sorted_vals = np.sort(values)
ecdf = np.arange(1, len(sorted_vals)+1) / len(sorted_vals)

plt.figure(figsize=(6, 4))
plt.plot(sorted_vals, 1 - ecdf, label='Empirical Tail (1-CDF)')
plt.yscale('log')
plt.xlabel('Value')
plt.ylabel('1 - CDF (Log Scale)')
plt.title('Empirical Tail Distribution')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Check number of exceedances
threshold = np.percentile(values, 95)  # 95th percentile as threshold
exceedances = values[values > threshold] - threshold
n_exceed = len(exceedances)

print(f"Threshold: {threshold:.2f}")
print(f"Number of exceedances: {n_exceed} ({n_exceed/len(values)*100:.2f}% of data)")
```

```python
# Fit GPD
params = genpareto.fit(exceedances)  # (shape, loc, scale)
shape, loc, scale = params
print(f"GPD fit parameters: shape={shape:.3f}, scale={scale:.3f}")
```

```python
# Goodness-of-Fit Test
ks_stat, ks_p = kstest(exceedances, 'genpareto', args=(shape, loc, scale))
print(f"KS test statistic: {ks_stat:.3f}, p-value: {ks_p:.3f}")
```

```python
# Histogram vs. Fitted GPD
x = np.linspace(min(exceedances), max(exceedances), 100)
pdf = genpareto.pdf(x, shape, loc, scale)

plt.figure(figsize=(6, 3))
sns.histplot(exceedances, kde=False, stat='density', bins=30, label='Exceedances')
plt.plot(x, pdf, 'r--', label='Fitted GPD')
plt.xlabel('Excess over threshold')
plt.ylabel('Density')
plt.title('GPD Fit to Exceedances')
plt.legend()
plt.tight_layout()
plt.show()
```

### 3. Change Point Detection 

```python
# Change point detection
model = rpt.Pelt(model="rbf").fit(values)
change_indices = model.predict(pen=10)

# Plot full signal with change points
plt.figure(figsize=(12, 4))
plt.plot(df.index, df["value"], label="Occupancy", linewidth=1)

for cp in change_indices[:-1]:  # Exclude final index (end of signal)
    plt.axvline(df.index[cp], color='red', linestyle='--', lw=1.2)

plt.title("Change Point Detection (PELT, RBF model)")
plt.xlabel("Time")
plt.ylabel("Occupancy")
plt.legend()
plt.tight_layout()
plt.show()
```

### 4. Would Moving Average Make the Residuals i.i.d.?

```python
# Prepare he dat 
df2 = pd.read_csv("../data/clean_data/post_gap_occupancy.csv", parse_dates=[0])
df2.columns = ["timestamp", "value"]

timestamps = pd.to_datetime(df2.iloc[:, 0])
raw = df2.iloc[:, 1]
raw.index = timestamps

# Define window sizes
windows = [50, 144, 288, 350]

# Set up plot grid
fig, axes = plt.subplots(len(windows), 2, figsize=(12, 3*len(windows)), sharex=False)

# Looop over windows
for i, window in enumerate(windows):
    ax_res, ax_acf = axes[i]

    # 4Compute rolling mean & residuals
    rolling_mean = raw.rolling(window=window, center=True).mean()
    residuals    = raw - rolling_mean

    # Plot residuals
    ax_res.plot(residuals, alpha=0.7)
    ax_res.set_title(f"Residuals (MA={window})")
    ax_res.set_ylabel("Value")

    # Plot ACF of residuals
    plot_acf(residuals.dropna(), lags=300, ax=ax_acf)
    ax_acf.set_title(f"ACF of Residuals (MA={window})")

plt.tight_layout()
plt.show()
```

#### Main Observations:

1. Windows 50 & 144 leave strong daily‐cycle autocorrelations in the residuals, so the series remains non-stationary.
2. Window 288 (one full day) effectively flattens most seasonality—residual ACF is near zero beyond short lags—while preserving the height of true spikes, failing to make series fully non-stationary.
3. Window 350 might to oversmoothing.
   
#### Implications for EVT-AD:
EVT assumes i.i.d. tails: only the MA=288 residuals come close to that, minimising bias in the estimated extreme-value distribution. Each window should be tested to see the sensitivity of the DSPOT to the size of the detrending as well as dealing with the remained non-stationarity.
